```python
x = wine_data.drop(['quality'], axis ='columns')
print(x)
```

```python
y = wine_data['quality'].apply(lambda y_value: 1 if y_value >= 7 else 0)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state =42)
```

```python
model = RandomForestClassifier(n_estimators=100)
```

```python
model.fit(X_train, y_train)
```

```python
X_training_predicting = model.predict(X_train)
Accuracy_score = accuracy_score(y_train, X_training_predicting)
print("Accuracy score :", Accuracy_score)
```

```python
#Testing the trained model using give input
input_data =(7.9, 0.6, 0.06,1.6, 0.069,	15.0, 59.0,	0.9964,	3.3, 0.46, 9.4)
input_data_array = np.asarray(input_data)
input_data_reshape = input_data_array.reshape(1,-1)
prediction = model.predict(input_data_reshape)

if prediction ==1:
    print("Good wine quality")
else:
    print("Bad wine quality")
```
